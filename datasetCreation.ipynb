```python
import cv2
import os
import time
import uuid
```

```python
 IMAGES_PATH = 'Tensorflow/workspace/images/cimages'
```

```python
labels = ['A']
#'zero','one', 'two','three','four','five','six', 'seven','eight', 'nine'
ni = 15
```

```python
for label in labels:
    os.mkdir('Tensorflow/workspace/images/cimages\\'+label)
    cap = cv2.VideoCapture(0)
    print('collecting image for {}'.format(label))
    time.sleep(6)
    for i in range(ni):
        ret, frame = cap.read()
        imgname = os.path.join(IMAGES_PATH, label, label+'_'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        cv2.imshow('frame',frame)
        time.sleep(5)
        
        key = cv2.waitKey(1)
        if key == 32:
            break
    cap.release()
cv2.destroyAllWindows()
cap.release()
```
